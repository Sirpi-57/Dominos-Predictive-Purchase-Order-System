#### 1.pizza_name_id:

```python
pizza_name_id_null = df_sales[df_sales['pizza_name_id'].isnull()]
print("Rows where pizza_name_id is null:")
print(pizza_name_id_null)
```

```python
# Display pizza_id, pizza_size, and pizza_name where pizza_name_id is null
null_pizza_details = pizza_name_id_null[['pizza_id', 'pizza_size', 'pizza_name']]
print("Details of rows with null pizza_name_id:")
print(null_pizza_details)
```

```python
# Identify the rows with null pizza_name_id
# First, identify rows with null pizza_name_id
null_rows = df_sales[df_sales['pizza_name_id'].isnull()]
print("Rows with null pizza_name_id:")
print(null_rows[['pizza_name', 'pizza_size']])

# Check if there are matching combinations with non-null pizza_name_id
matching_ids = df_sales[df_sales['pizza_name_id'].notnull()].groupby(['pizza_name', 'pizza_size'])['pizza_name_id'].first()
print("\nAvailable pizza_name_id combinations:")
print(matching_ids)

# Fill null values
for index, row in null_rows.iterrows():
    try:
        df_sales.at[index, 'pizza_name_id'] = matching_ids.loc[(row['pizza_name'], row['pizza_size'])]
    except KeyError:
        print(f"No matching ID found for {row['pizza_name']} in size {row['pizza_size']}")
```

```python
print("Null values in Sales Table:")
print(df_sales.isnull().sum())
```

#### total_price:

```python
total_price_null = df_sales[df_sales['total_price'].isnull()]
print("Rows where total_price is null:")
print(total_price_null)
```

```python
# Fill total_price where it is null using quantity and unit_price
df_sales['total_price'] = df_sales['total_price'].fillna(df_sales['quantity'] * df_sales['unit_price'])
```

```python
print("Null values in Sales Table:")
print(df_sales.isnull().sum())
```

#### pizza_category:

```python
pizza_category_null = df_sales[df_sales['pizza_category'].isnull()]
print("Rows where pizza_category  is null:")
print(pizza_category_null)
```

```python
# Display pizza_id, pizza_size, and pizza_name where pizza_name_id is null
null_pizza_details = pizza_category_null[['pizza_category', 'pizza_ingredients', 'pizza_name']]
print("Details of rows with null pizza_category:")
print(null_pizza_details)
```

```python
# Find unique combinations of pizza_name and pizza_category
unique_combinations = df_sales[['pizza_name', 'pizza_category']].drop_duplicates()
print(unique_combinations)
```

```python
# Step 1: Create a mapping of pizza_name to pizza_category (excluding null categories)
pizza_category_mapping = df_sales[['pizza_name', 'pizza_category']].dropna().drop_duplicates()
pizza_category_mapping = pizza_category_mapping.set_index('pizza_name')['pizza_category'].to_dict()

# Step 2: Fill null pizza_category using the mapping
df_sales['pizza_category'] = df_sales['pizza_category'].fillna(df_sales['pizza_name'].map(pizza_category_mapping))

# Step 3: Display rows where pizza_category was previously null
updated_rows = df_sales.loc[df_sales['pizza_name'].isin([
    'The Mexicana Pizza', 'The Brie Carre Pizza', 'The Pepper Salami Pizza',
    'The Napolitana Pizza', 'The Spinach Pesto Pizza', 'The Hawaiian Pizza',
    'The Sicilian Pizza', 'The Spicy Italian Pizza', 'The Barbecue Chicken Pizza',
    'The Chicken Alfredo Pizza', 'The Pepperoni Pizza', 'The Greek Pizza',
    'The Spinach and Feta Pizza', 'The Italian Capocollo Pizza',
    'The Five Cheese Pizza', 'The Prosciutto and Arugula Pizza'
])]
print("Updated rows with filled pizza_category:")
print(updated_rows[['pizza_name', 'pizza_category']])
```

```python
print("Null values in Sales Table:")
print(df_sales.isnull().sum())
```

#### pizza_ingredients:

```python
# Step 1: Create a mapping of pizza_name to pizza_category (excluding null categories)
pizza_ingredients_mapping = df_sales[['pizza_name', 'pizza_ingredients']].dropna().drop_duplicates()
pizza_ingredients_mapping = pizza_ingredients_mapping.set_index('pizza_name')['pizza_ingredients'].to_dict()

# Step 2: Fill null pizza_category using the mapping
df_sales['pizza_ingredients'] = df_sales['pizza_ingredients'].fillna(df_sales['pizza_name'].map(pizza_ingredients_mapping))
```

#### pizza_name:

```python
pizza_name_null = df_sales[df_sales['pizza_name'].isnull()]
print("Rows where pizza_name  is null:")
print(pizza_name_null)
```

```python
# Step 1: Identify rows with null pizza_name
pizza_name_null = df_sales[df_sales['pizza_name'].isnull()]
print("Rows where pizza_name is null:")
print(pizza_name_null[['pizza_name_id', 'pizza_name']])

# Step 2: Identify the pizza_name_id for which we need to fill pizza_name
pizza_name_id_null = pizza_name_null['pizza_name_id'].unique()
print("\nUnique pizza_name_id values with null pizza_name:")
print(pizza_name_id_null)

# Step 3: Look for similar rows based on pizza_name_id
similar_combinations = df_sales[df_sales['pizza_name_id'].isin(pizza_name_id_null)]
print("\nRows with similar pizza_name_id (with non-null pizza_name):")
print(similar_combinations[['pizza_name_id', 'pizza_name']])

```

```python
# Fill null pizza_name based on similar pizza_name_id
for id_value in pizza_name_id_null:
   # Find the most common pizza name for this pizza_name_id
   most_common_name = df_sales[df_sales['pizza_name_id'] == id_value]['pizza_name'].mode().values[0]
   
   # Fill null pizza_name where pizza_name_id matches
   df_sales.loc[(df_sales['pizza_name_id'] == id_value) & (df_sales['pizza_name'].isnull()), 'pizza_name'] = most_common_name

# Verify the changes
print("Rows where pizza_name was null:")
print(df_sales[df_sales['pizza_name'].isnull()])
```

```python
import pandas as pd

# Sample DataFrame creation (use your actual df_ing DataFrame here)
# df_ing = pd.read_csv('your_dominos_data.csv')  # Load your dataset

# Step 1: Define a function to extract the pizza size based on the last character(s)
def extract_pizza_size(pizza_name_id):
    if 'xl' in pizza_name_id:  # Check for xl
        return 'xl'
    elif 'xxl' in pizza_name_id:  # Check for xxl
        return 'xxl'
    elif pizza_name_id[-1] == 'l':  # Check for large size by the last letter 'l'
        return 'l'
    elif pizza_name_id[-1] == 'm':  # Check for medium size by the last letter 'm'
        return 'm'
    elif pizza_name_id[-1] == 's':  # Check for small size by the last letter 's'
        return 's'
    else:
        return 'unknown'  # In case of some unexpected value

# Step 2: Apply the function to create a new column 'pizza_size'
df_ing['pizza_size'] = df_ing['pizza_name_id'].apply(extract_pizza_size)

# Step 3: Calculate the sum quantity for each pizza_name_id
sum_qty = df_ing.groupby('pizza_name_id')['Items_Qty_In_Grams'].sum().reset_index()
sum_qty.rename(columns={'Items_Qty_In_Grams': 'sum_qty'}, inplace=True)

# Step 4: Merge the sum_qty with the original DataFrame
df_ing = pd.merge(df_ing, sum_qty, on='pizza_name_id', how='left')

# Step 5: Verify the results
print(df_ing[['pizza_name_id', 'pizza_size', 'sum_qty']].head())
```

### Features for EDA:

```python
# Convert order_date and order_time to datetime format
df_sales['order_date'] = pd.to_datetime(df_sales['order_date'])
df_sales['order_time'] = pd.to_timedelta(df_sales['order_time'].astype(str))
```

```python
# Extract year, month, week, day of the week, and weekday/weekend
df_sales['year'] = df_sales['order_date'].dt.year
df_sales['month'] = df_sales['order_date'].dt.month
df_sales['week'] = df_sales['order_date'].dt.isocalendar().week  # Week of year
df_sales['day_of_week'] = df_sales['order_date'].dt.day_name()  # Name of day
```

```python
# Weekend or Weekday
df_sales['is_weekend'] = df_sales['day_of_week'].isin(['Saturday', 'Sunday'])
df_sales['day_type'] = df_sales['is_weekend'].apply(lambda x: 'Weekend' if x else 'Weekday')
```

```python
#Time-based Features: working hours vs non-working hours
def working_hours(hour):
    if 9 <= hour < 18:
        return 'Working Hours'
    else:
        return 'Non-Working Hours'

def time_period(hour):
    if 5 <= hour < 12:
        return 'Morning'
    elif 12 <= hour < 17:
        return 'Afternoon'
    elif 17 <= hour < 21:
        return 'Evening'
    elif 21 <= hour < 24:
        return 'Night'
    else:
        return 'Midnight'

# Extract hour from order_time
df_sales['hour'] = df_sales['order_time'].dt.seconds // 3600  # Convert timedelta to hours

df_sales['working_hours'] = df_sales['hour'].apply(working_hours)
df_sales['time_period'] = df_sales['hour'].apply(time_period)
```

```python
# Drop helper column 'hour' if not needed
df_sales = df_sales.drop(columns=['hour'])

# Final Output
print(df_sales)
```

```python
# Map month numbers to month names
month_mapping = {
    1: 'January', 2: 'February', 3: 'March', 4: 'April',
    5: 'May', 6: 'June', 7: 'July', 8: 'August',
    9: 'September', 10: 'October', 11: 'November', 12: 'December'
}

# Replace month numbers with month names
df_sales['month'] = df_sales['month'].map(month_mapping)

# Verify the output
print(df_sales[['order_date', 'month']])
```
